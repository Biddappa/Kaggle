```python
train_features = train_data.drop(['Survived', 'PassengerId', 'age_group'], axis=1)
train_target = train_data["Survived"]
X = np.array(train_features)
y = np.array(train_target)
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
feature_coloumn = ['Pclass','SibSp','Parch','Fare','CabinBool', 'Family_size','Title','Age','Sex',
                   'Embarked']
```

```python
gkf = KFold(n_splits=5, shuffle=True, random_state=42)#.split(X=X_train, y=y_train)
df_metric_data_all = pd.DataFrame(columns=['Index','Valid','Predict'])
for train_index, test_index in gkf.split(X):
    train_X, valid_X = X[train_index], X[test_index]
    train_y, valid_y = y[train_index], y[test_index]
    param_grid = {
               #'boosting_type': ['gbdt'],
                'max_depth': [4, 6, 8, 10, 12, None],
                'n_estimators': [100, 300, 500, 700,900]}
              #'learning_rate': [0.1, 0.03, 0.01]}
    xgbclas = xgb(random_state=42)#min_samples_leaf=2, max_features='auto',
    gsearch = GridSearchCV(estimator=xgbclas, param_grid=param_grid, cv=3, n_jobs=-1)
    lgbreg_model = gsearch.fit(train_X, train_y)
    #lgbreg_model = xgbclas.fit(train_X, train_y)
    predict = lgbreg_model.predict(valid_X)
    zip_result = zip(train_index, valid_y, predict)
    zip_tuple = list(zip_result)
    df_metric_data = pd.DataFrame(zip_tuple, columns=['Index','Valid','Predict'])
    #print(df_metric_data)
    df_metric_data_all = df_metric_data_all.append(df_metric_data, ignore_index=True)
    #Valid_List, Predict_List = zip(*zip_list)
    #Valid_List_all.append(Valid_List)
    #Predict_List_all.append(Predict_List)

print(df_metric_data_all)
```
